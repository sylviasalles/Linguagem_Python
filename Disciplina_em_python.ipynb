```python
import numpy as np #Biblioteca que carrega números
import matplotlib.pyplot as plt #Biblioteca cria gráficos
import time #Biblioteca de tempo, hora, duração 
from IPython import display

def gera_poligono(nl,ang):
    angr = ang*np.pi/180
    d = np.linspace(0, 2*np.pi, nl+1) + angr
    
    x=np.sin(d)
    y=np.cos(d)
    
    return x,y

x, y = gera_poligono(3, 15)
x2, y2 = gera_poligono(5, 10)

plt.plot(x,y)
plt.plot(x2,y2)
plt.axis('off')
plt.axis('equal')
plt.show()
```

```python
a360 = np.arange(0, 360)
for a in a360:
    x, y = gera_poligono(4, a)
    x2, y2 = gera_poligono(3, -a)
    x2 = x2*2
    y2 = y2*2
    
    plt.figure(figsize=(10,10))
    plt.plot(x,y)
    plt.plot(x2,y2)
    plt.axis('off')
    plt.axis('equal')
    plt.xlim(-2.3, 2.3)
    plt.ylim(-2.3, 2.3)
    #plt.show()
    
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(.1)
    plt.close()
```

### Como montar uma função

```python
x=[1,2,3]
conta=0
for num in x:
    conta=conta+1
    print(conta)
print('conta total=',conta)
```

```python
def contagem(lista):
    conta=0
    for _ in lista:
        conta=conta+1
    
    return conta
```

```python
#Código para fazer uma contagem
x=[1,4,6] 
contagem=0
for num in x:
    contagem=contagem+1
print(contagem)
```

```python
#Função para fazer contagem
def tuti(lista):
    contagem=0
    for num in lista:
        contagem=contagem+1
    return contagem
```

```python
#Código para fazer soma
x=[1,4,6]
soma=0
for num in x:
    soma+= num
print(soma)
```

```python
#Função para calcular a soma
def pitoca(lista):
    soma=0
    for num in lista:
        soma+= num
    return soma
```

```python
#Teste, usando função contagem e soma
preta=[9,8]
lola=pitoca(preta)
kika=tuti(preta)
lola/kika
```

```python
#Função para calcular a média
def luna(lista):
    a=tuti(lista)
    b=pitoca(lista)
    c=b/a
    return c
```

```python
#Teste, calculo da média
d=[6,7]
e=luna(d)
print(e)
```

```python
#Código para calcular Desvio Padrão
x=[5,2]
numero_de_elementos=tuti(x)
media_dos_elementos=luna(x)
soma_dos_elementos=0
for i in range(numero_de_elementos):
    diferenca=x[i]-media_dos_elementos
    diferenca_ao_quadrado=diferenca**2
    soma_dos_elementos+=diferenca_ao_quadrado
variancia=soma_dos_elementos/(numero_de_elementos-1)
desvio_padrao=variancia**0.5
```

```python
#Função para calcular Desvio Padrão
def belinha(lista):
    numero_de_elementos=tuti(lista)
    media_dos_elementos=luna(lista)
    soma_dos_elementos=0
    for i in range(numero_de_elementos):
        diferenca=lista[i]-media_dos_elementos
        diferenca_ao_quadrado=diferenca**2
        soma_dos_elementos+=diferenca_ao_quadrado
    variancia=soma_dos_elementos/(numero_de_elementos-1)
    desvio_padrao=variancia**0.5
    return desvio_padrao
```

```python
#Teste, calculo do desvio padrão
gordo=[5,6]
g=belinha(gordo)
print(g)
```
